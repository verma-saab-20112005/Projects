```python
data["date_ist"]= pd.to_datetime(data["date_ist"])
```

```python
data["time_ist"]= data["time_ist"].astype("string")
data["Hours"]= data["time_ist"].str.split(":", expand=True)[0].astype("int64")
data.drop(columns=["time_ist"], inplace=True)
```
