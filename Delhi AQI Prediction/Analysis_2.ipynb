```python
data["date_ist"]= pd.to_datetime(data["date_ist"])
```

```python
data['datetime'] = data['date_ist'] + pd.to_timedelta(data['time_ist'].astype(str))
```

```python
data["datetime"].dtype
```

```python
data["year"]= data["datetime"].dt.year
```

```python
data["month"]= data["datetime"].dt.month
```

```python
data["day"]= data["datetime"].dt.day
```

```python
data["Hours"]= data["datetime"].dt.hour
```

```python
data["Minutes"]= data["datetime"].dt.minute
```

```python
data["Seconds"]= data["datetime"].dt.second
```

```python
data= data.drop(["date_ist", "time_ist", "datetime"], axis=1)
```
