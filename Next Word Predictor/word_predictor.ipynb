```python
from tensorflow import keras
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer= Tokenizer()
```

```python
tokenizer.fit_on_texts([data])
```

```python
len(tokenizer.word_index)
```

```python
tokenizer.document_count
```

```python
input_data=[]
for sentence in data.split("\n"):
    token_data= tokenizer.texts_to_sequences([sentence])[0]
    
    for i in range(1, len(token_data)):
        input_data.append(token_data[:i+1])
```

```python
input_data
```

```python
from keras.utils import pad_sequences
```

```python
padded_data= pad_sequences(input_data, padding="pre")
```

```python
padded_data.shape
```

```python
x= padded_data[:,:-1]
y= padded_data[:,-1]
```

```python
x.shape, y.shape
```

```python
x
```

```python
from keras.utils import to_categorical
```

```python
y= to_categorical(y, num_classes=293)
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
from keras.layers import Dense, Flatten, Input, SimpleRNN, LSTM, Dropout, BatchNormalization, Embedding
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from keras.regularizers import L2
```

```python
model= Sequential([
    Input(shape=(62,)),
    Embedding(input_dim=293, output_dim=100),
    LSTM(200, return_sequences=True),
    LSTM(150, return_sequences=True),
    LSTM(100),
    Dense(293, activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(
    optimizer="Adam",
    loss="categorical_crossentropy",
    metrics= ["accuracy"]
)
```

```python
callback= EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=2,
    mode="auto",
    baseline= None,
    restore_best_weights=True,
    verbose=1
)
```

```python
history= model.fit(x_train, y_train, epochs=200)
```

```python
text="An oily water"
for i in range(15):
    token_text= tokenizer.texts_to_sequences([text])
    pad_text= pad_sequences(token_text, padding="pre", maxlen=62)
    pos= np.argmax(model.predict(pad_text))
    for word,index in tokenizer.word_index.items():
        if index==pos:
            text= text+" "+word
            print(text)
```
