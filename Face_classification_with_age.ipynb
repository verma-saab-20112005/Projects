```python
import pandas as pd
import numpy as np
```

```python
import tensorflow as tf
import keras
```

```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
import os
```

```python
folder_path= r"D:\Data Science\Campusx\Non-Linear NN\Face Classification with age\UTKFace"
```

```python
img_path=[]
gender=[]
age=[]

for file in os.listdir(folder_path):
    parts = file.split("_")
    
    age.append(int(parts[0]))
    gender.append(int(parts[1]))
    img_path.append(file)
```

```python
data= pd.DataFrame({"Age":age, "Gender":gender, "IMG":img_path})
```

```python
data
```

```python
train_data = data.sample(frac=1,random_state=42).iloc[:20000]
test_data = data.sample(frac=1,random_state=42).iloc[20000:]
```

```python
train_data, test_data
```

```python
train_datagen= ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode= "nearest"
)

test_datagen= ImageDataGenerator(rescale=1./255)
```

```python
train_generator= train_datagen.flow_from_dataframe(
    train_data,
    directory= folder_path,
    x_col= "IMG",
    y_col=["Age", "Gender"],
    target_size= (200,200),
    class_mode="multi_output"
)
```

```python
test_generator= test_datagen.flow_from_dataframe(
    test_data,
    directory= folder_path,
    x_col= "IMG",
    y_col=["Age", "Gender"],
    target_size= (200,200),
    class_mode="multi_output"
)
```

```python
from keras.applications.vgg16 import VGG16
```

```python
vggnet= VGG16(
    weights="imagenet",
    include_top=False,
    input_shape=(200,200,3)
)
```

```python
vggnet.trainable= False
```

```python
from keras.models import Model
from keras.layers import Dense, Flatten, Input, Dropout, BatchNormalization
from keras.regularizers import L2
from keras.callbacks import EarlyStopping
```

```python
o1= vggnet.layers[-1].output
```

```python
flatten= Flatten()(o1)
```

```python
x1= Dense(128, activation="relu", kernel_regularizer=L2(0.001))(flatten)
x2= Dense(64, activation="relu", kernel_regularizer=L2(0.001))(x1)
x3= Dense(16, activation="relu", kernel_regularizer=L2(0.001))(x2)
x4= Dense(4, activation="relu", kernel_regularizer=L2(0.001))(x3)
output1= Dense(1, activation="linear", name="Age", kernel_regularizer=L2(0.001))(x4)
```

```python
y1= Dense(128, activation="relu", kernel_regularizer=L2(0.001))(flatten)
y2= Dense(64, activation="relu", kernel_regularizer=L2(0.001))(y1)
y3= Dense(16, activation="relu", kernel_regularizer=L2(0.001))(y2)
y4= Dense(4, activation="relu", kernel_regularizer=L2(0.001))(y3)
output2= Dense(1, activation="sigmoid", name="Gender", kernel_regularizer=L2(0.001))(y4)
```

```python
model= Model(inputs= vggnet.input, outputs= [output1, output2])
```

```python
model.compile(
    optimizer="Adam",
    loss={
        "Age":"mean_squared_error",
        "Gender":"binary_crossentropy"
    },
    metrics={
        "Age":["mse"],
        "Gender":["accuracy"]
    }
)
```

```python
callback= EarlyStopping(
    monitor="val_loss",
    min_delta=0.001,
    patience=2,
    mode="auto",
    baseline=None,
    restore_best_weights=True,
    verbose=1
)
```

```python
def split_outputs(generator):
    for x, y in generator:
        yield x, {
            "Age": y[0],
            "Gender": y[1]
        }
```

```python
history= model.fit(
    split_outputs(train_generator),
    epochs=25,
    validation_data=split_outputs(test_generator),
    callbacks=callback
)
```
